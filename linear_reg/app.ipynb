```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ornek_sayisi = 20
x = np.arange(-ornek_sayisi / 2, ornek_sayisi / 2)
print(x)
```

```python
w = 2
y = w * x + np.random.normal(0, 1, ornek_sayisi)
print(y)
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel("X", fontsize=18)
plt.ylabel("Y", fontsize=18)
plt.scatter(x, y)
plt.show()
```

```python
def toplam_hata(gercek_degerler, tahmin_degerleri):
    ornek_sayisi = len(gercek_degerler)
    farklarin_karesi = np.square(gercek_degerler - tahmin_degerleri)
    toplam_fark_kareleri = np.sum(farklarin_karesi)
    return toplam_fark_kareleri / (2 * ornek_sayisi)

print(toplam_hata(np.array([1, 2, 8]), np.array([3, 5, 4])))
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel(r"${x}$", fontsize=18)
plt.ylabel(r"${y}$", fontsize=18)
plt.scatter(x, y)

fd = {"size": 16}

theta = -1
y_hat = theta * x
plt.plot(x, y_hat)
plt.text(x[-1] + 1, y_hat[-1], f"y = -1 * x  hata  {toplam_hata(y, y_hat):.2f}", fontdict=fd, va="center")

theta = 0
y_hat = theta * x
plt.plot(x, y_hat)
plt.text(x[-1] + 1, y_hat[-1], f"y = 0 * x  hata  {toplam_hata(y, y_hat):.2f}", fontdict=fd, va="center")

theta = 1
y_hat = theta * x
plt.plot(x, y_hat)
plt.text(x[-1] + 1, y_hat[-1], f"y = 1 * x  hata  {toplam_hata(y, y_hat):.2f}", fontdict=fd, va="center")

theta = 2
y_hat = theta * x
plt.plot(x, y_hat)
plt.text(x[-1] + 1, y_hat[-1], f"y = 2 * x  hata  {toplam_hata(y, y_hat):.2f}", fontdict=fd, va="center")

theta = 3
y_hat = theta * x
plt.plot(x, y_hat)
plt.text(x[-1] + 1, y_hat[-1], f"y = 3 * x  hata  {toplam_hata(y, y_hat):.2f}", fontdict=fd, va="center")

plt.xlim(-12, ornek_sayisi + 8)
plt.tight_layout()
plt.savefig("hatalar.pdf")
plt.show()
```

```python
theta_degerleri = []
toplam_hatalar = []

for theta in np.linspace(-5, 10, 31):
    y_hat = theta * x
    hata = toplam_hata(y, y_hat)
    
    theta_degerleri.append(theta)
    toplam_hatalar.append(hata)

    print(f"theta = {theta} \t toplam hata = {hata:.2f}")
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel(r"$\theta_1$", fontsize=18)
plt.ylabel("Hata (Loss)", fontsize=18)
plt.scatter(theta_degerleri, toplam_hatalar)
plt.tight_layout()
plt.savefig("loss.pdf")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel(r"$\theta_1$", fontsize=18)
plt.ylabel("Hata (Loss)", fontsize=18)
plt.xlim(-5, 12)
plt.ylim(-5, 1200)
plt.scatter(theta_degerleri[25], toplam_hatalar[25])
plt.tight_layout()
plt.savefig("loss_single.pdf")
plt.show()
```

```python
ornek_sayisi = 10
x = np.arange(-ornek_sayisi / 2, ornek_sayisi / 2)

w = 1.5
y = w * x + np.random.normal(0, 1, ornek_sayisi)

plt.figure(figsize=(12, 8))
plt.xlabel(r"${x}$", fontsize=18)
plt.ylabel(r"${y}$", fontsize=18)

fd = {"size": 16}

theta = 0.5
y_hat = theta * x
plt.plot(x, y_hat, linewidth=6, color="orange")
plt.text(x[-1] + 0.5, y_hat[-1], r"$y = 0.5 x$", fontdict=fd, va="center")

plt.plot((x, x), ([i for (i, j) in zip(y, y_hat)], [j for (i, j) in zip(y, y_hat)]), c='gray', linestyle='dashed')

plt.scatter(x, y, s=200)

plt.xlim(-6, 6)
plt.tight_layout()
plt.savefig("hatalar_isaretli.pdf")
plt.show()
```

```python
theta_degerleri = []
toplam_hatalar = []

for theta in np.linspace(-3, 6.1, 100):
    y_hat = theta * x
    hata = toplam_hata(y, y_hat)
    
    theta_degerleri.append(theta)
    toplam_hatalar.append(hata)

plt.figure(figsize=(6, 6))
plt.xlabel(r"$\theta_1$", fontsize=18)
plt.ylabel(r"$J(\theta_1)$", fontsize=18)
plt.plot(theta_degerleri, toplam_hatalar)
plt.tight_layout()
plt.savefig("loss_u.pdf")
plt.show()
```
